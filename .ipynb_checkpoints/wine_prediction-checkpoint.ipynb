```python
def compute_threshold(series, no_of_deviations): 
    std_ = series.std() 
    threshold_max = series.mean() + no_of_deviations*std_ 
    threshold_min = series.mean() - no_of_deviations*std_ 

    return threshold_max, threshold_min 

def remove_outliers(frame, col, threshold_max, threshold_min, drop_threshold): 
    
    outliers = frame[(frame[col] > threshold_max) | (frame[col] < threshold_min)]
    no_outliers = len(outliers) 
    if no_outliers/len(frame) < drop_threshold: 
        new_frame = frame.drop(outliers.index)
        
    return new_frame
```

```python
threshold_max, threshold_min = compute_threshold(wine_copy['residual sugar'], 3)
col = 'residual sugar' 
drop_threshold = 0.1 
series = wine_copy
new_series = remove_outliers(series, col, threshold_max, threshold_min, drop_threshold)
```

```python
wine_copy = wine_copy.drop(['wine_type', 'quality_cat'], axis=1)
wine_copy.head()
```

```python
for col in wine_copy: 
    series = wine_copy[col]
    deviations = 3
    drop = 0.1
    threshold_max, threshold_min = compute_threshold(series, deviations)
    wine_copy = remove_outliers(wine_copy, col, threshold_max, threshold_min, drop) 
```
